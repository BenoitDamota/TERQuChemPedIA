## Préparation des données et entraînement des modèles DIST_REL_CC_02, DIST_REL_CH_02 et DIST_REL_OH_02

Le modèle DIST_REL_C_02 donnant de bons résultats, nous allons généraliser la méthode aux liaisions carbone-hydrogène et oxygène-hydrogène. Nous allons également entraîner le modèle DIST_REL_CC_02, qui est identique au modèle DIST_REL_C_02, à la différence qu'il s'entraînera sur plus d'exemples.

La différence entre les modèles DIST_REL_XY_01 et les modèles DIST_REL_XY_02 est que l'on va maintenant leur donner une information moindre mais plus pertinente. On cherche toujours à prédire les distances entre des liaisons XY, mais on va seulement donner les informations concernant les atomes au voisinage de la liaison. Formellement, nous allons donner les mêmes informations, mais uniquement pour les atomes à moins de 2 angströms de la liaison.
